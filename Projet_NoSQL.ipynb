```python
# Barre de recherche pour les noms des espaces verts

espacesVertsRechercher = espacesVerts.find({"fields.nom": { "$regex": 'place' }}, {"_id" : 0, "fields.nom": 1, "fields.arrondissement": 1})

print(espacesVertsRechercher.count())

for x in espacesVertsRechercher:
    print(x)
```

```python
# Trier les parc par:   arondissement et proportion de vegation (Tri a multiple paramètres)
# ici nous recherchons les espaces verts du type Jardinets décoratifs et Décorations sur la voie publique dans le 18e arrondissement

espacesVertsTrier = espacesVerts.find({"fields.type": {"$in": ["Jardinets décoratifs", "Décorations sur la voie publique"]}, "fields.proportion_vegetation_haute" : {"$lt":2}}, {'_id': 0, "fields.nom": 1, "fields.arrondissement": 1, 'fields.type': 1, 'fields.proportion_vegetation_haute': 1})

print(espacesVertsTrier.count())
for x in espacesVertsTrier:
    print(x)
```

```python
# Tri par rapport au nombre de velib dispo dans les stations

VelibAumoinsNbr = velib.find({"fields.numbikesavailable" : {"$gte":46}},{"_id" : 0,"fields.name" : 1,"fields.numbikesavailable" : 1, "fields.coordonnees_geo" : 1,})

print("\n Station avec Nbr de velib dispos : \n")
print(VelibAumoinsNbr.count())
for x in VelibAumoinsNbr:
  print(x)

# Tri par rapport au nombre de place pour déposer les vélib dispo dans les stations

placeAuMoinsNbr = velib.find({"fields.numdocksavailable" : {"$gte":50}},{"_id" : 0,"fields.name" : 1,"fields.numbikesavailable" : 1, "fields.coordonnees_geo" : 1,})

print("\n Sation avec Nbr de place libre pour déposer un vélib : \n")
print(placeAuMoinsNbr.count())
for x in placeAuMoinsNbr:
  print(x)
```

```python
# requete complexe a multiple paramètres
# tous les espaces verts qui ne sont pas dans le 1er, 4e ou 12e arrondissement, qui est ouvert le dimanche et qui a une proportion de vegetation haute entre 2 compris et 2,5 exclu

requeteComplexe = espacesVerts.find({"fields.arrondissement": {"$not": {"$in": [1, 4, 12]}}, "fields.horaires_dimanche": { "$exists": 1 }, "fields.proportion_vegetation_haute": {"$gte":2,"$lt":2.5}}, {"_id" : 0, "fields.nom": 1, "fields.arrondissement": 1})

for x in requeteComplexe:
    print(x)
```
